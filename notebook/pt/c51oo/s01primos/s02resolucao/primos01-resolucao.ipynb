# Números Primos

```c
#include <stdio.h>

int main() {
    int numero = 4,
        flag=0;

    for (int i = 2; i < numero; i++)
        if (numero % i == 0)
            flag = 1;

    if (flag == 0)
        printf("Este numero eh primo");
    else
        printf("Não eh primo");


    return 0;
}
```

```c
#include <stdio.h>

int main() {
    int n = 10,
        flag;

    int numero = 1;
    int t = 0;
    while (t < n) {
        flag = 0;
        for (int i = 2; i < numero; i++)
            if (numero % i == 0)
                flag = 1;

        if (flag == 0) {
            printf("%d\n", numero);
            t++;
        }
        
        numero++;
    }

    return 0;
}
```

# Exercício

Construa um gerador de números primos seguindo a lógica de um Tipo Abstrato de Dados.

```c
#include <stdio.h>

typedef struct {
    int n;
    int corrente;
    int primoCorrente;
} SequenciaPrimos;

SequenciaPrimos novaSequencia(int n) {
    SequenciaPrimos umaSequencia;
    umaSequencia.n = n;
    umaSequencia.corrente = 0;
    umaSequencia.primoCorrente = 0;
    return umaSequencia;
}

int testaPrimo(int numero) {
   int flag = 1;
   for (int i = 2; i < numero; i++)
      if (numero % i == 0)
          flag = 0;
   return flag;
}

int proximoPrimo(SequenciaPrimos *sequencia) {
    int retorno = 0;
    if (sequencia->corrente < sequencia->n) {
        sequencia->corrente++;
        int flag;
        do {
            sequencia->primoCorrente++;
            flag = testaPrimo(sequencia->primoCorrente);
        } while (flag == 0);
        retorno = sequencia->primoCorrente;
    }
    return retorno;
}

int main() {
    SequenciaPrimos sequencia = novaSequencia(10);

    int primo = proximoPrimo(&sequencia);
    
    while (primo > 0) {
        printf("%d\n", primo);
        primo = proximoPrimo(&sequencia);
    }

    return 0;
}
```
